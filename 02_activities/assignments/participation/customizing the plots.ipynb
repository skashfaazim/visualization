```python
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])


# Font styles for titles and labels
title_font = {'family': 'serif', 'color': 'darkblue', 'size': 14}
label_font = {'family': 'sans-serif', 'color': 'gray', 'size': 10}


fig, (ax1, ax2) = plt.subplots(ncols = 2,
                               nrows =1,
                               figsize=(7, 3))
ax1.scatter(x1, y1, marker='*', color='darkred', edgecolor='black', alpha=0.8)
ax1.set_title("Random Pirate Stats", fontdict=title_font)
ax1.set_xlabel("Index", fontdict=label_font)
ax1.set_ylabel("Power Level", fontdict=label_font)
ax1.grid(True, linestyle='--', linewidth=0.5, color='lightgray')
ax1.annotate("Max Power", xy=(x1[np.argmax(y1)], max(y1)), 
             xytext=(35, 70),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=9)
# Second subplot: bar chart
colors = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99', '#FFD700']
ax2.bar(x2, y2, color=colors)
ax2.set_title("Pirate Crew Strengths", fontdict=title_font)
ax2.set_ylabel("Strength Score", fontdict=label_font)
ax2.set_xlabel("Character", fontdict=label_font)
ax2.grid(axis='y', linestyle=':', linewidth=0.7, color='gray')
ax2.bar_label(ax2.containers[0], fontsize=9)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                     ['ax2', 'ax3']],
                                     figsize = (7,4),
                                     layout='constrained')
someaxes['ax1'].scatter(x1, y1)
someaxes['ax2'].bar(x2, y2)
someaxes['ax3'].plot(x1, y1)
someaxes['ax1'].set_xlabel('A Big Label', fontsize = 18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
```
